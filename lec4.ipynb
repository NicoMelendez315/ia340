```python
df['year']=pd.to_datetime(df['built_in'], format = "%Y" )

df.head()
```

```python
df['unit_price'] = df['price']/df['area']
df.head()
```

```python
df = df.dropna()
df.info()
```

```python
df['price'].mean()
```

```python
avg_price_per_type = df.groupby('house_type')['price'].mean()
```

```python
avg_price_per_type.head()
```

```python
avg_price_per_type.plot.bar()
```
